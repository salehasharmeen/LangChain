```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Annotated, List
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import JsonOutputParser, StrOutputParser
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain.agents import create_openai_tools_agent, AgentExecutor
import operator

# 1. Define the Graph State (Adding user_feedback to the state)
class GraphState(TypedDict):
    """Represents the state of our graph.

    Attributes:
        topic: The initial research topic from the user.
        keywords: Keywords generated by the topic scoping agent.
        research_questions: Research questions generated by the topic scoping agent.
        search_results: The raw search results from the literature search agent.
        review: The final synthesized literature review.
        user_feedback: User's response to the search results.
    """
    topic: str
    keywords: List[str]
    research_questions: List[str]
    search_results: str
    review: str
    user_feedback: str

# 2. Define the Nodes
def run_topic_scoping(state: GraphState):
    """Runs the topic scoping agent to get keywords and questions."""
    print("---RUNNING TOPIC SCOPING NODE---")
    topic_scoping_chain = ChatPromptTemplate.from_messages(
        [
            ("system", """You are a helpful assistant for academic research. Your role is to analyze a broad research topic provided by the user and break it down into more specific, actionable research questions and relevant keywords.
            Follow these instructions:
            1. Analyze the user's research topic.
            2. Propose 3-5 specific, well-defined research questions that could be used for a deep literature review.
            3. Generate a list of at least 10 relevant keywords and search phrases that would be useful for a literature search.
            4. Format your output as a JSON object with two keys: "research_questions" (a list of strings) and "keywords" (a list of strings)."""),
            ("user", "{topic}"),
        ]
    ) | ChatOpenAI(model="gpt-4o-mini", temperature=0.1) | JsonOutputParser()

    response = topic_scoping_chain.invoke({"topic": state["topic"]})

    return {
        "keywords": response["keywords"],
        "research_questions": response["research_questions"]
    }

def run_literature_search(state: GraphState):
    """Runs the literature search agent using keywords from the state."""
    print("---RUNNING LITERATURE SEARCH NODE---")
    llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1)
    search = TavilySearchResults(max_results=5)

    search_agent = create_openai_tools_agent(
        llm=llm,
        tools=[search],
        prompt=ChatPromptTemplate.from_messages(
            [
                ("system", """You are a specialized research assistant. Use the provided `tavily_search_results` tool to find academic papers and articles.
                Use the keywords and research questions provided. Your final answer must be the raw search results content, not a summary.
                Your goal is to find information that is specific to the user's research questions. If a previous search failed, try a new combination of keywords."""),
                ("placeholder", "{chat_history}"),
                ("user", "{input}"),
                ("placeholder", "{agent_scratchpad}"),
            ]
        )
    )
    search_agent_executor = AgentExecutor(agent=search_agent, tools=[search])

    # Check for user feedback from previous loop
    if "user_feedback" in state and state["user_feedback"] == "no":
        print("---REFINING SEARCH BASED ON USER FEEDBACK---")
        # Example of how to refine the query: ask the user for a more specific term
        refinement_query = input("Please specify a more precise term for the next search: ")
        search_query = "academic research on " + refinement_query + " " + " ".join(state["keywords"])
    else:
        search_query = "academic research on " + " ".join(state["keywords"][:5])

    response = search_agent_executor.invoke({"input": search_query})

    return {"search_results": response["output"], "user_feedback": "initial"}

def get_user_feedback(state: GraphState):
    """Asks the user if the search results are satisfactory."""
    print("---WAITING FOR USER FEEDBACK---")
    print("\n------------------")
    print("Here are the search results:")
    print(state["search_results"])
    print("------------------\n")

    feedback = input("Are these search results satisfactory for a literature review? (yes/no): ")
    return {"user_feedback": feedback.lower()}

def run_literature_review(state: GraphState):
    """Synthesizes the search results into a polished literature review."""
    print("---RUNNING LITERATURE REVIEW NODE---")
    review_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", """You are a highly skilled academic literature review specialist. Your task is to synthesize raw content from multiple research papers and articles into a coherent, professional, and structured literature review. The content you receive will be from a search tool, so it may include titles, URLs, and text. Ensure you reference these details in your review.
            Your review should:
            - Start with a brief introduction.
            - Have a section for "Key Findings" that summarizes the main points.
            - Include a section for "Identified Themes or Gaps."
            - Be written in a formal, academic tone."""),
            ("user", "Here is the raw content from several research papers:\n\n{text_content}"),
        ]
    )
    review_chain = review_prompt | ChatOpenAI(model="gpt-4o-mini", temperature=0.1) | StrOutputParser()

    return {"review": review_chain.invoke({"text_content": state["search_results"]})}

def route_to_next_step(state: GraphState):
    """Decides the next step based on user feedback."""
    if state["user_feedback"] == "yes":
        print("---ROUTER: User is satisfied, proceeding to review.---")
        return "review"
    else:
        print("---ROUTER: User is not satisfied, looping back to search.---")
        return "search"

# 3. Build the Graph
workflow = StateGraph(GraphState)

workflow.add_node("topic_scoping", run_topic_scoping)
workflow.add_node("literature_search", run_literature_search)
workflow.add_node("user_feedback", get_user_feedback)
workflow.add_node("literature_review", run_literature_review)

workflow.add_edge("topic_scoping", "literature_search")
workflow.add_edge("literature_search", "user_feedback")

workflow.add_conditional_edges(
    "user_feedback",
    route_to_next_step,
    {
        "review": "literature_review",
        "search": "literature_search",
    }
)

workflow.add_edge("literature_review", END)
workflow.set_entry_point("topic_scoping")

app = workflow.compile()

# 4. Run the full, multi-step application as a stream
user_topic = input("Enter a research topic to generate a full review: ")
inputs = {"topic": user_topic}
for s in app.stream(inputs):
    print(s)
```

```python
%%writefile app.py
import streamlit as st
from langgraph.graph import StateGraph, END
from typing import TypedDict, Annotated, List
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import JsonOutputParser, StrOutputParser
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain.agents import create_openai_tools_agent, AgentExecutor
import os

# Set page config
st.set_page_config(
    page_title="Academic Research Agent",
    page_icon="📚",
    layout="wide"
)

# Custom CSS for better styling
st.markdown("""
<style>
    .main-header {
        font-size: 2.5rem;
        color: #1f77b4;
        text-align: center;
        margin-bottom: 2rem;
    }
    .section-header {
        font-size: 1.8rem;
        color: #2c3e50;
        margin-top: 2rem;
        margin-bottom: 1rem;
    }
    .result-box {
        background-color: #f8f9fa;
        border: 1px solid #e9ecef;
        border-radius: 8px;
        padding: 1rem;
        margin: 1rem 0;
    }
    .feedback-buttons {
        display: flex;
        gap: 1rem;
        margin: 1rem 0;
    }
</style>
""", unsafe_allow_html=True)

# Title and description
st.markdown('<h1 class="main-header">📚 Academic Research Agent</h1>', unsafe_allow_html=True)
st.markdown("""
This tool helps you conduct comprehensive literature reviews by:
1. Generating research questions and keywords from your topic
2. Searching academic literature
3. Synthesizing findings into a professional review
""")

# Initialize session state
if 'research_state' not in st.session_state:
    st.session_state.research_state = None
if 'current_step' not in st.session_state:
    st.session_state.current_step = 'start'
if 'search_results' not in st.session_state:
    st.session_state.search_results = None
if 'user_feedback' not in st.session_state:
    st.session_state.user_feedback = None

# 1. Define the Graph State
class GraphState(TypedDict):
    topic: str
    keywords: List[str]
    research_questions: List[str]
    search_results: str
    review: str
    user_feedback: str

# 2. Define the Nodes
def run_topic_scoping(state: GraphState):
    """Runs the topic scoping agent to get keywords and questions."""
    st.session_state.current_step = 'topic_scoping'

    topic_scoping_chain = ChatPromptTemplate.from_messages(
        [
            ("system", """You are a helpful assistant for academic research. Your role is to analyze a broad research topic provided by the user and break it down into more specific, actionable research questions and relevant keywords.
            Follow these instructions:
            1. Analyze the user's research topic.
            2. Propose 3-5 specific, well-defined research questions that could be used for a deep literature review.
            3. Generate a list of at least 10 relevant keywords and search phrases that would be useful for a literature search.
            4. Format your output as a JSON object with two keys: "research_questions" (a list of strings) and "keywords" (a list of strings)."""),
            ("user", "{topic}"),
        ]
    ) | ChatOpenAI(model="gpt-4o-mini", temperature=0.1) | JsonOutputParser()

    response = topic_scoping_chain.invoke({"topic": state["topic"]})
    return {
        "keywords": response["keywords"],
        "research_questions": response["research_questions"]
    }

def run_literature_search(state: GraphState):
    """Runs the literature search agent using keywords from the state."""
    st.session_state.current_step = 'literature_search'

    llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1)
    search = TavilySearchResults(max_results=5)

    search_agent = create_openai_tools_agent(
        llm=llm,
        tools=[search],
        prompt=ChatPromptTemplate.from_messages(
            [
                ("system", """You are a specialized research assistant. Use the provided `tavily_search_results` tool to find academic papers and articles.
                Use the keywords and research questions provided. Your final answer must be the raw search results content, not a summary.
                Your goal is to find information that is specific to the user's research questions. If a previous search failed, try a new combination of keywords."""),
                ("placeholder", "{chat_history}"),
                ("user", "{input}"),
                ("placeholder", "{agent_scratchpad}"),
            ]
        )
    )
    search_agent_executor = AgentExecutor(agent=search_agent, tools=[search])

    if "user_feedback" in state and state["user_feedback"] == "no":
        # For Streamlit, we'll handle refinement differently
        search_query = "academic research on " + " ".join(state["keywords"][:5])
    else:
        search_query = "academic research on " + " ".join(state["keywords"][:5])

    response = search_agent_executor.invoke({"input": search_query})
    st.session_state.search_results = response["output"]
    return {"search_results": response["output"], "user_feedback": "initial"}

def run_literature_review(state: GraphState):
    """Synthesizes the search results into a polished literature review."""
    st.session_state.current_step = 'literature_review'

    review_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", """You are a highly skilled academic literature review specialist. Your task is to synthesize raw content from multiple research papers and articles into a coherent, professional, and structured literature review. The content you receive will be from a search tool, so it may include titles, URLs, and text. Ensure you reference these details in your review.
            Your review should:
            - Start with a brief introduction.
            - Have a section for "Key Findings" that summarizes the main points.
            - Include a section for "Identified Themes or Gaps."
            - Be written in a formal, academic tone."""),
            ("user", "Here is the raw content from several research papers:\n\n{text_content}"),
        ]
    )
    review_chain = review_prompt | ChatOpenAI(model="gpt-4o-mini", temperature=0.5) | StrOutputParser()

    return {"review": review_chain.invoke({"text_content": state["search_results"]})}

# 3. Build the Graph
def build_workflow():
    workflow = StateGraph(GraphState)

    workflow.add_node("topic_scoping", run_topic_scoping)
    workflow.add_node("literature_search", run_literature_search)
    workflow.add_node("literature_review", run_literature_review)

    workflow.add_edge("topic_scoping", "literature_search")
    workflow.add_edge("literature_search", "literature_review")
    workflow.add_edge("literature_review", END)

    workflow.set_entry_point("topic_scoping")
    return workflow.compile()

# Main app logic
def main():
    # Check if API keys are already set in environment
    openai_api_key = os.environ.get("OPENAI_API_KEY")
    tavily_api_key = os.environ.get("TAVILY_API_KEY")

    # If keys are not in environment, show input fields
    if not openai_api_key or not tavily_api_key:
        st.sidebar.header("Configuration")
        openai_api_key = st.sidebar.text_input("OpenAI API Key", type="password")
        tavily_api_key = st.sidebar.text_input("Tavily API Key", type="password")

        if not openai_api_key or not tavily_api_key:
            st.warning("Please enter both API keys to continue")
            return

    # Set the environment variables (this ensures they're available to all components)
    os.environ["OPENAI_API_KEY"] = openai_api_key
    os.environ["TAVILY_API_KEY"] = tavily_api_key

    # Research topic input
    if st.session_state.current_step == 'start':
        topic = st.text_input("Enter your research topic:")
        if st.button("Start Research") and topic:
            st.session_state.research_state = {"topic": topic}
            st.session_state.current_step = 'processing'
            st.rerun()

    # Processing steps
    if st.session_state.current_step == 'processing':
        with st.spinner("Processing your research request..."):
            try:
                app = build_workflow()
                inputs = {"topic": st.session_state.research_state["topic"]}

                for s in app.stream(inputs):
                    if 'topic_scoping' in s:
                        st.session_state.research_state.update(s['topic_scoping'])
                    elif 'literature_search' in s:
                        st.session_state.research_state.update(s['literature_search'])
                    elif 'literature_review' in s:
                        st.session_state.research_state.update(s['literature_review'])

                st.session_state.current_step = 'results'
                st.rerun()

            except Exception as e:
                st.error(f"An error occurred: {str(e)}")
                st.session_state.current_step = 'start'

    # Display results
    if st.session_state.current_step == 'results' and st.session_state.research_state:
        st.markdown('<h2 class="section-header">Research Results</h2>', unsafe_allow_html=True)

        # Research Questions
        st.subheader("Research Questions")
        for i, question in enumerate(st.session_state.research_state.get("research_questions", []), 1):
            st.write(f"{i}. {question}")

        # Keywords
        st.subheader("Keywords")
        st.write(", ".join(st.session_state.research_state.get("keywords", [])))

        # Search Results
        st.subheader("Search Results")
        st.text_area("Raw Search Results",
                    st.session_state.research_state.get("search_results", ""),
                    height=200)

        # Literature Review
        st.subheader("Literature Review")
        st.markdown(st.session_state.research_state.get("review", ""))

        # Reset button
        if st.button("Start New Research"):
            st.session_state.clear()
            st.rerun()

if __name__ == "__main__":
    main()
```
